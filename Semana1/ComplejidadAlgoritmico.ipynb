```python
import numpy as np
import matplotlib.pyplot as plt
import time
from google.colab import files

# -----------------------------
# Tamaños del problema
# -----------------------------
N_values = [100,200,400,600,800,1000]

tiempo_lineal = []
tiempo_nlogn = []
tiempo_cuadratico = []

# -----------------------------
# Medición de tiempos
# -----------------------------
for N in N_values:

    # Algoritmo ~ N
    inicio = time.perf_counter()

    s = 0
    for i in range(N):
        s += i

    fin = time.perf_counter()
    tiempo_lineal.append(fin - inicio)


    # Algoritmo ~ N log N
    inicio = time.perf_counter()

    s = 0
    k = int(np.log(N)) + 1

    for i in range(N):
        for j in range(k):
            s += j

    fin = time.perf_counter()
    tiempo_nlogn.append(fin - inicio)


    # Algoritmo ~ N^2
    inicio = time.perf_counter()

    s = 0
    for i in range(N):
        for j in range(N):
            s += j

    fin = time.perf_counter()
    tiempo_cuadratico.append(fin - inicio)


# -----------------------------
# Gráfica de resultados
# -----------------------------
plt.figure(figsize=(7,5))

plt.plot(N_values,tiempo_lineal,'o-',label=r"$N$")
plt.plot(N_values,tiempo_nlogn,'o-',label=r"$N\log N$")
plt.plot(N_values,tiempo_cuadratico,'o-',label=r"$N^2$")

plt.xlabel("Tamaño del problema N")
plt.ylabel("Tiempo de ejecución (s)")
plt.title("Comparación de tiempos de ejecución de algoritmos")

plt.yscale('log')

plt.legend()

# -----------------------------
# Guardar figura
# -----------------------------
plt.savefig("Comparacion_algoritmos.pdf",dpi=300,bbox_inches='tight')

plt.show()

# -----------------------------
# Descargar automáticamente
# -----------------------------
files.download("Comparacion_algoritmos.pdf")
```

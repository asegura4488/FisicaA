```python
"""
Comparación de escalamiento de algoritmos

Este script compara el tiempo de ejecución de dos algoritmos
con crecimiento aproximado N log N y N^2. Luego calcula
la diferencia entre ambos tiempos y genera una gráfica.

Autor: Curso Física A
"""

import numpy as np
import matplotlib.pyplot as plt
import time

# -------------------------------------
# Tamaños del problema
# -------------------------------------

N_values = [100, 200, 400, 600, 800, 1000]

tiempo_nlogn = []
tiempo_cuadratico = []

# -------------------------------------
# Medición de tiempos
# -------------------------------------

for N in N_values:

    # Algoritmo ~ N log N
    inicio = time.perf_counter()

    s = 0
    k = int(np.log(N)) + 1

    for i in range(N):
        for j in range(k):
            s += j

    fin = time.perf_counter()

    tiempo_nlogn.append(fin - inicio)


    # Algoritmo ~ N^2
    inicio = time.perf_counter()

    s = 0
    for i in range(N):
        for j in range(N):
            s += j

    fin = time.perf_counter()

    tiempo_cuadratico.append(fin - inicio)


# -------------------------------------
# Conversión a arrays
# -------------------------------------

tiempo_nlogn = np.array(tiempo_nlogn)
tiempo_cuadratico = np.array(tiempo_cuadratico)

# Diferencia de tiempos
diferencia = tiempo_cuadratico - tiempo_nlogn


# -------------------------------------
# Gráfica
# -------------------------------------

plt.figure(figsize=(7,5))

plt.plot(N_values, diferencia, 'o-', label=r"$T_{N^2} - T_{N\log N}$")

plt.xlabel("Tamaño del problema N")
plt.ylabel("Diferencia de tiempo (s)")
plt.title("Diferencia de tiempos entre algoritmos")

plt.legend()

# -------------------------------------
# Guardar figura
# -------------------------------------

nombre_figura = "diferencia_algoritmos.pdf"
plt.savefig(nombre_figura, dpi=300, bbox_inches='tight')

plt.show()

print("Figura guardada como:", nombre_figura)


# -------------------------------------
# Descarga automática si se usa Colab
# -------------------------------------

try:
    from google.colab import files
    files.download(nombre_figura)
except:
    pass
```

```python
from google.colab import drive
drive.mount('/content/drive')
from google.colab import files

import numpy as np
import pandas as pd
import os

# Ruta absoluta (IMPORTANTE)
carpeta = "/content/drive/MyDrive/Datos_Lab1"
os.makedirs(carpeta, exist_ok=True)

ruta = os.path.join(carpeta, "datos_parabola.csv")

# Crear datos
x = np.linspace(-5, 5, 30)
y = 2*x**2 + 3*x + 1 + np.random.normal(0, 5, size=len(x))

df = pd.DataFrame({"x": x, "y": y})
df.to_csv(ruta, index=False)

print("Archivo guardado en:", ruta)

# Descargar archivo al computador
files.download(ruta)
```

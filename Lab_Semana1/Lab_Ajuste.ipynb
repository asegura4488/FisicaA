```python
from google.colab import drive
drive.mount('/content/drive')
from google.colab import files

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os

# =====================================================
# LEER DATOS DESDE LA CARPETA
# =====================================================

carpeta = "/content/drive/MyDrive/Proyecto_FisicaA/FisicaA/Data/"
archivo = "Datos_parabola.csv"

ruta = os.path.join(carpeta, archivo)

df = pd.read_csv(ruta)

x = df["x"].values
y = df["y"].values

# =====================================================
# AJUSTE CUADRATICO
# =====================================================

coef = np.polyfit(x, y, 2)
a, b, c = coef

print("Coeficientes del ajuste:")
print(f"a = {a:.4f}")
print(f"b = {b:.4f}")
print(f"c = {c:.4f}")

# =====================================================
# GRAFICA
# =====================================================

x_fit = np.linspace(min(x), max(x), 200)
y_fit = a*x_fit**2 + b*x_fit + c

plt.figure(figsize=(6,5))
plt.scatter(x, y, color='k' ,label="Datos experimentales")
plt.plot(x_fit, y_fit, color='r' ,label="Ajuste cuadrático", linewidth=2)

plt.xlabel("x")
plt.ylabel("y")
plt.title("Ajuste parabólico desde archivo CSV")
plt.legend()

# Guardar en la misma carpeta
ruta_fig = os.path.join(carpeta, "Ajuste_Parabola.pdf")
plt.savefig(ruta_fig, dpi=300, bbox_inches="tight")

plt.show()

# Descargar figura
files.download(ruta_fig)
```

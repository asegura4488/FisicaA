```python
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files

# -----------------------------
# Parametros del problema
# -----------------------------
x_real = 10
sigma_ruido = 0.5

N_values = np.array([5, 10, 20, 50, 100, 200, 500, 1000])

errores_promedio = []
teoria = sigma_ruido / np.sqrt(N_values)

# -----------------------------
# Simulacion de mediciones
# -----------------------------
for N in N_values:

    mediciones = x_real + np.random.normal(0, sigma_ruido, N)

    promedio = np.mean(mediciones)

    # Error del promedio estimado numericamente
    error_prom = np.std(mediciones, ddof=1) / np.sqrt(N)

    errores_promedio.append(error_prom)

    print("N =", N)
    print("Promedio =", promedio)
    print("Error del promedio =", error_prom)
    print("-------------------------")

errores_promedio = np.array(errores_promedio)

# -----------------------------
# Grafica
# -----------------------------
plt.figure(figsize=(8,5))

plt.plot(N_values, errores_promedio, 'o-', label="Simulacion")
plt.plot(N_values, teoria, 's--', label=r"$\sigma/\sqrt{N}$")

plt.xlabel("Numero de mediciones N")
plt.ylabel("Incertidumbre del promedio")
plt.title("Comparacion entre simulacion y teoria para el error del promedio")

plt.legend()

# -----------------------------
# Guardar figura
# -----------------------------
plt.savefig("Actividad_6_1_error_promedio_teoria.pdf", dpi=300, bbox_inches="tight")

plt.show()

# -----------------------------
# Descargar automaticamente
# -----------------------------
files.download("Actividad_6_1_error_promedio_teoria.pdf")
```

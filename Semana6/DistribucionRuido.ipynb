```python
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files

# -----------------------------
# Parámetros del problema
# -----------------------------
x_real = 10          # valor real
N = 100              # número de mediciones
sigma_ruido = 0.5    # desviación del ruido

# -----------------------------
# Simulación de mediciones
# -----------------------------
ruido = np.random.normal(0, sigma_ruido, N)
mediciones = x_real + ruido

# -----------------------------
# Estadísticos
# -----------------------------
promedio = np.mean(mediciones)
sigma = np.std(mediciones, ddof=1)

print("Valor promedio:", promedio)
print("Desviación estándar:", sigma)

# -----------------------------
# Gráfica
# -----------------------------
plt.figure(figsize=(8,5))

plt.hist(mediciones, bins=15)

plt.xlabel("Valor medido")
plt.ylabel("Frecuencia")
plt.title("Distribución de mediciones experimentales")

# -----------------------------
# Guardar figura
# -----------------------------
plt.savefig("Distribucion_mediciones.pdf", dpi=300, bbox_inches="tight")

plt.show()

# -----------------------------
# Descargar automáticamente
# -----------------------------
files.download("Distribucion_mediciones.pdf")
```

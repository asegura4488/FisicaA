```python
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files

# -----------------------------
# Parámetros medidos
# -----------------------------
L = 5.0
sigma_L = 0.1

W = 3.0
sigma_W = 0.1

N = 10000

# -----------------------------
# Simulación Monte Carlo
# -----------------------------
L_sim = np.random.normal(L, sigma_L, N)
W_sim = np.random.normal(W, sigma_W, N)

A = L_sim * W_sim

# -----------------------------
# Resultados
# -----------------------------
A_prom = np.mean(A)
sigma_A = np.std(A, ddof=1)

print("Área promedio:", A_prom)
print("Incertidumbre del área:", sigma_A)

# -----------------------------
# Gráfica
# -----------------------------
plt.figure(figsize=(8,5))

plt.hist(A, bins=30)

plt.xlabel("Área (cm²)")
plt.ylabel("Frecuencia")
plt.title("Distribución del área calculada")

# -----------------------------
# Guardar figura
# -----------------------------
plt.savefig("Propagacion_area.pdf", dpi=300, bbox_inches="tight")

plt.show()

# -----------------------------
# Descargar automáticamente
# -----------------------------
files.download("Propagacion_area.pdf")
```

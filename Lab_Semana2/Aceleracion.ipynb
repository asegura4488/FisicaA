```python
from google.colab import drive
drive.mount('/content/drive')
from google.colab import files

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os

# =====================================================
# LECTURA DE DATOS DE PHYPHOX DESDE GOOGLE DRIVE
# =====================================================

# Cambia este nombre por el archivo exportado desde Phyphox
carpeta = "/content/drive/MyDrive/Proyecto_FisicaA/Data/"
archivo = "Lab2/Aceleracion_rawdata.csv"

ruta = os.path.join(carpeta, archivo)

# Leer archivo CSV
df = pd.read_csv(ruta)

print("Primeras filas del archivo:")
print(df.head())

print("\nColumnas disponibles:")
print(df.columns)

# =====================================================
# EXTRAER COLUMNAS
# =====================================================
# Este bloque asume columnas con nombres:
# time, ax, ay, az
# Si tu archivo tiene otros nombres, ajusta estas etiquetas.

t = df["Time (s)"].values
ax = df["Acceleration x (m/s^2)"].values
ay = df["Acceleration y (m/s^2)"].values
az = df["Acceleration z (m/s^2)"].values

# =====================================================
# MAGNITUD TOTAL DE LA ACELERACION
# =====================================================

a_mag = np.sqrt(ax**2 + ay**2 + az**2)

# =====================================================
# GRAFICA ax vs tiempo
# =====================================================

plt.figure(figsize=(6, 5))
plt.plot(t, ax)
plt.xlabel("t")
plt.ylabel("a_x")
plt.title("Componente a_x en función del tiempo")
ruta_fig1 = os.path.join(carpeta, "Phyphox_ax_vs_t.pdf")
plt.savefig(ruta_fig1, dpi=300, bbox_inches="tight")
plt.show()
files.download(ruta_fig1)

# =====================================================
# GRAFICA ay vs tiempo
# =====================================================

plt.figure(figsize=(6, 5))
plt.plot(t, ay)
plt.xlabel("t")
plt.ylabel("a_y")
plt.title("Componente a_y en función del tiempo")
ruta_fig2 = os.path.join(carpeta, "Phyphox_ay_vs_t.pdf")
plt.savefig(ruta_fig2, dpi=300, bbox_inches="tight")
plt.show()
files.download(ruta_fig2)

# =====================================================
# GRAFICA az vs tiempo
# =====================================================

plt.figure(figsize=(6, 5))
plt.plot(t, az)
plt.xlabel("t")
plt.ylabel("a_z")
plt.title("Componente a_z en función del tiempo")
ruta_fig3 = os.path.join(carpeta, "Phyphox_az_vs_t.pdf")
plt.savefig(ruta_fig3, dpi=300, bbox_inches="tight")
plt.show()
files.download(ruta_fig3)

# =====================================================
# GRAFICA MAGNITUD DE LA ACELERACION
# =====================================================

plt.figure(figsize=(6, 5))
plt.plot(t, a_mag)
plt.xlabel("t")
plt.ylabel("|a|")
plt.title("Magnitud de la aceleración en función del tiempo")
ruta_fig4 = os.path.join(carpeta, "Phyphox_amagnitud_vs_t.pdf")
plt.savefig(ruta_fig4, dpi=300, bbox_inches="tight")
plt.show()
files.download(ruta_fig4)

# =====================================================
# RESUMEN NUMERICO
# =====================================================

print("\nResumen:")
print(f"Promedio de a_x = {np.mean(ax):.4f}")
print(f"Promedio de a_y = {np.mean(ay):.4f}")
print(f"Promedio de a_z = {np.mean(az):.4f}")
print(f"Promedio de |a| = {np.mean(a_mag):.4f} m/s^2")
print(f"Desviación estándar de |a| = {np.std(a_mag):.4f} m/s^2")
```

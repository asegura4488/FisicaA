```python
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files

# =====================================================
# ACTIVIDAD 13.1
# EXPLORACION GUIADA DE ONDAS VIAJERAS
# =====================================================

# -----------------------------------------------------
# Dominio espacial
# -----------------------------------------------------
x = np.linspace(0, 10, 400)

# -----------------------------------------------------
# Funcion de onda gaussiana viajera
# u(x,t) = exp(-(x - c t - x0)^2)
# -----------------------------------------------------
def onda_gaussiana(x, t, c=1.0, x0=3.0):
    return np.exp(-(x - c * t - x0)**2)

# -----------------------------------------------------
# Funcion con dos picos iniciales
# -----------------------------------------------------
def onda_doble(x, t, c=1.0):
    return np.exp(-(x - c * t - 3.0)**2) + 0.5 * np.exp(-(x - c * t - 6.0)**2)

# =====================================================
# PARTE 1: PROPAGACION TEMPORAL DE UNA ONDA
# =====================================================

c = 1.0
tiempos = [0, 1, 2, 3, 4]

plt.figure(figsize=(8, 5))

for t in tiempos:
    u = onda_gaussiana(x, t, c=c, x0=3.0)
    plt.plot(x, u, label=f"t={t}")

plt.xlabel("x")
plt.ylabel("u(x,t)")
plt.title("Actividad 13.1: propagacion de una onda gaussiana")
plt.legend()
plt.tight_layout()
plt.savefig("Actividad_Onda_Tiempo.pdf", dpi=300, bbox_inches="tight")
plt.show()

files.download("Actividad_Onda_Tiempo.pdf")

# =====================================================
# PARTE 2: COMPARACION DE VELOCIDADES
# =====================================================

velocidades = [0.5, 1.0, 2.0]
t_fijo = 2.0

plt.figure(figsize=(8, 5))

for c in velocidades:
    u = onda_gaussiana(x, t=t_fijo, c=c, x0=3.0)
    plt.plot(x, u, label=f"c={c}")

plt.xlabel("x")
plt.ylabel("u(x,t)")
plt.title(f"Comparacion de velocidades de propagacion en t={t_fijo}")
plt.legend()
plt.tight_layout()
plt.savefig("Actividad_Onda_Velocidades.pdf", dpi=300, bbox_inches="tight")
plt.show()

files.download("Actividad_Onda_Velocidades.pdf")

# =====================================================
# PARTE 3: ONDA CON DOS PICOS INICIALES
# =====================================================

plt.figure(figsize=(8, 5))

for t in tiempos:
    u = onda_doble(x, t, c=1.0)
    plt.plot(x, u, label=f"t={t}")

plt.xlabel("x")
plt.ylabel("u(x,t)")
plt.title("Propagacion de una perturbacion con dos picos")
plt.legend()
plt.tight_layout()
plt.savefig("Actividad_Onda_Doble.pdf", dpi=300, bbox_inches="tight")
plt.show()

files.download("Actividad_Onda_Doble.pdf")

# =====================================================
# PARTE 4: SEGUIMIENTO DEL MAXIMO DE LA ONDA
# =====================================================

c = 1.0
tiempos_max = np.linspace(0, 4, 50)
x_maximos = []

for t in tiempos_max:
    u = onda_gaussiana(x, t, c=c, x0=3.0)
    indice_max = np.argmax(u)
    x_maximos.append(x[indice_max])

plt.figure(figsize=(8, 5))
plt.plot(tiempos_max, x_maximos)
plt.xlabel("t")
plt.ylabel("Posicion del maximo")
plt.title("Seguimiento del maximo de la onda")
plt.tight_layout()
plt.savefig("Actividad_Onda_Maximo.pdf", dpi=300, bbox_inches="tight")
plt.show()

files.download("Actividad_Onda_Maximo.pdf")

# =====================================================
# RESUMEN NUMERICO
# =====================================================

print("===================================")
print("ACTIVIDAD 13.1: EXPLORACION DE ONDAS")
print("===================================")
print("Se generaron las siguientes figuras:")
print("1. Actividad_Onda_Tiempo.pdf")
print("2. Actividad_Onda_Velocidades.pdf")
print("3. Actividad_Onda_Doble.pdf")
print("4. Actividad_Onda_Maximo.pdf")
print("")
print("Interpretacion sugerida:")
print("- La onda conserva su forma al propagarse.")
print("- La velocidad c controla cuanto se desplaza en el tiempo.")
print("- Multiples picos iniciales se propagan manteniendo su estructura.")
print("- La posicion del maximo cambia linealmente con el tiempo.")
```

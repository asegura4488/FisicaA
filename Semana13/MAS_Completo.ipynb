```python
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files

t = np.linspace(0, 10, 1000)

A = 1
omega = 2

x = A * np.cos(omega * t)
v = -A * omega * np.sin(omega * t)

plt.plot(t, x, label="Posición")
plt.plot(t, v, label="Velocidad")
plt.xlabel("t")
plt.ylabel("Magnitud")
plt.title("Movimiento armónico simple")
plt.legend()

plt.savefig("MAS_Completo.pdf")
plt.show()

files.download("MAS_Completo.pdf")
```

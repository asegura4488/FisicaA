```python
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files

t = np.linspace(0, 10, 1000)

A = 1
omega = 2
m = 1
k = omega**2

x = A * np.cos(omega * t)
v = -A * omega * np.sin(omega * t)

Ep = 0.5 * k * x**2
Ec = 0.5 * m * v**2
Et = Ep + Ec

plt.plot(t, Ep, label="Ep")
plt.plot(t, Ec, label="Ec")
plt.plot(t, Et, label="Total")

plt.xlabel("t")
plt.ylabel("Energia")
plt.title("Energía en el oscilador")
plt.legend()

plt.savefig("MAS_Energia.pdf")
plt.show()

files.download("MAS_Energia.pdf")
```

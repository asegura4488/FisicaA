```python
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files

# -----------------------------
# Parámetros del problema
# -----------------------------
C0 = 12      # capacidad inicial (Ah)
I = 2        # corriente (A)

# Tiempo total de descarga
t_total = C0 / I

# Arreglo de tiempo
t = np.linspace(0, t_total, 300)

# Capacidad restante
C = C0 - I*t

# -----------------------------
# Porcentajes de capacidad
# -----------------------------
C50 = 0.5*C0
C20 = 0.2*C0
C10 = 0.1*C0

# tiempos correspondientes
t50 = (C0 - C50)/I
t20 = (C0 - C20)/I
t10 = (C0 - C10)/I

print("Tiempo al 50%:",t50,"h")
print("Tiempo al 20%:",t20,"h")
print("Tiempo al 10%:",t10,"h")

# -----------------------------
# Gráfica
# -----------------------------
plt.figure(figsize=(8,5))

plt.plot(t,C,color='k',label="Capacidad restante")

# puntos importantes
plt.scatter([t50],[C50],color='r',label='$t_{50}$')
plt.scatter([t20],[C20],color='b',label='$t_{20}$')
plt.scatter([t10],[C10],color='g',label='$t_{10}$')

# líneas de referencia
#plt.axhline(C50,label="50%")
#plt.axhline(C20,label="20%")
#plt.axhline(C10,label="10%")

#plt.axvline(t50)
#plt.axvline(t20)
#plt.axvline(t10)

plt.xlabel("Tiempo (h)")
plt.ylabel("Capacidad restante (Ah)")
plt.title("Descarga de una batería")

plt.xlim(0,t_total)
plt.ylim(0,C0)

plt.legend()

# -----------------------------
# Guardar figura
# -----------------------------
plt.savefig("Descarga_bateria_porcentajes.pdf",dpi=300,bbox_inches="tight")

plt.show()

# -----------------------------
# Descargar figura
# -----------------------------
files.download("Descarga_bateria_porcentajes.pdf")
```

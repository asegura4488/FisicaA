```python
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files

# -----------------------------
# Parámetros físicos
# -----------------------------
C0 = 12            # capacidad inicial (Ah)
k0 = 0.20          # h^{-1}
T0 = 20            # temperatura de referencia (°C)
alpha = 0.02       # 1/°C

temperaturas = [20, 30, 40]

# -----------------------------
# Parámetros numéricos
# -----------------------------
dt = 0.01
t_max = 15
t = np.arange(0, t_max, dt)

# -----------------------------
# Gráfica
# -----------------------------
plt.figure(figsize=(8,5))

for T in temperaturas:

    k = k0 * (1 + alpha * (T - T0))

    C = np.zeros(len(t))
    C[0] = C0

    # Euler
    for n in range(len(t)-1):
        C[n+1] = C[n] - k * C[n] * dt

    plt.plot(t, C, label=f"T = {T} °C")

plt.xlabel("Tiempo (h)")
plt.ylabel("Capacidad restante (Ah)")
plt.title("Descarga de una batería para distintas temperaturas")

plt.xlim(0, t_max)
plt.ylim(0, C0)

plt.legend()

# -----------------------------
# Guardar figura
# -----------------------------
plt.savefig("Actividad_5_1_bateria_temperatura.pdf", dpi=300, bbox_inches="tight")

plt.show()

# -----------------------------
# Descargar automáticamente
# -----------------------------
files.download("Actividad_5_1_bateria_temperatura.pdf")
```

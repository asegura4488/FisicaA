```python
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files

# -----------------------------
# Parámetros físicos
# -----------------------------
C0 = 12        # capacidad inicial (Ah)
k = 0.25       # tasa de consumo

# tiempo de simulación
t_max = 10
dt = 0.01

# arreglo de tiempo
t = np.arange(0, t_max, dt)

# arreglo de capacidad
C = np.zeros(len(t))
C[0] = C0

# -----------------------------
# Integración con Euler
# -----------------------------
for n in range(len(t)-1):

    I = k * C[n]        # corriente dependiente de la capacidad

    C[n+1] = C[n] - I * dt

# -----------------------------
# Solución exacta
# -----------------------------
C_exact = C0 * np.exp(-k*t)

# -----------------------------
# Gráfica
# -----------------------------
plt.figure(figsize=(8,5))

plt.plot(t, C_exact, label="Solución exacta")
plt.plot(t, C, '--', label="Euler")

plt.xlabel("Tiempo (h)")
plt.ylabel("Capacidad restante (Ah)")
plt.title("Descarga de batería con corriente dependiente de la capacidad")

plt.xlim(0, t_max)
plt.ylim(0, C0)

plt.legend()

# -----------------------------
# Guardar figura
# -----------------------------
plt.savefig("Descarga_bateria_Euler.pdf", dpi=300, bbox_inches="tight")

plt.show()

# -----------------------------
# Descargar automáticamente
# -----------------------------
files.download("Descarga_bateria_Euler.pdf")
```

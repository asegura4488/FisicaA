```python
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files

# =====================================================
# INTERFERENCIA EN 2D (DOS FUENTES)
# =====================================================

# Malla
N = 250
x = np.linspace(-5, 5, N)
y = np.linspace(-5, 5, N)
X, Y = np.meshgrid(x, y)

# Parametros
k = 3.0
omega = 2.0
t = 0.0
d = 1.5

# Distancias a las fuentes
r1 = np.sqrt((X + d)**2 + Y**2)
r2 = np.sqrt((X - d)**2 + Y**2)

# Evitar singularidades
r1[r1 == 0] = 1e-6
r2[r2 == 0] = 1e-6

# Ondas
u1 = np.sin(k*r1 - omega*t)
u2 = np.sin(k*r2 - omega*t)

# Interferencia
u = u1 + u2

# -----------------------------------------------------
# GRAFICA
# -----------------------------------------------------
plt.figure(figsize=(6,6))

plt.imshow(u, extent=[-5,5,-5,5])
plt.xlabel("x")
plt.ylabel("y")
plt.title("Patrón de interferencia en 2D")

plt.tight_layout()
plt.savefig("Interferencia_2D.pdf", dpi=300, bbox_inches="tight")
plt.show()

files.download("Interferencia_2D.pdf")
```

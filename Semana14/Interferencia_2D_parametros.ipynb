```python
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files

# =====================================================
# EXPLORACION DE PARAMETROS EN INTERFERENCIA 2D
# =====================================================

N = 200
x = np.linspace(-5, 5, N)
y = np.linspace(-5, 5, N)
X, Y = np.meshgrid(x, y)

k_values = [2.0, 4.0]
d_values = [1.0, 2.0]

fig, axs = plt.subplots(2, 2, figsize=(8, 8))

for i, k in enumerate(k_values):
    for j, d in enumerate(d_values):

        r1 = np.sqrt((X + d)**2 + Y**2)
        r2 = np.sqrt((X - d)**2 + Y**2)

        r1[r1 == 0] = 1e-6
        r2[r2 == 0] = 1e-6

        u = np.sin(k * r1) + np.sin(k * r2)

        axs[i, j].imshow(u, extent=[-5, 5, -5, 5])
        axs[i, j].set_title(f"k={k}, d={d}")

plt.tight_layout()
plt.savefig("Interferencia_2D_parametros.pdf", dpi=300, bbox_inches="tight")
plt.show()

files.download("Interferencia_2D_parametros.pdf")
```

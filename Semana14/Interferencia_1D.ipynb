```python
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files

x = np.linspace(0, 10, 500)

k = 2.0
omega = 2.0
t = 0.0

phi = np.pi / 2

u1 = np.cos(k*x - omega*t)
u2 = np.cos(k*x - omega*t + phi)

u = u1 + u2

plt.plot(x, u1, label="u1")
plt.plot(x, u2, label="u2")
plt.plot(x, u, label="suma", linewidth=2)

plt.xlabel("x")
plt.ylabel("u")
plt.title("Interferencia en 1D")
plt.legend()

plt.savefig("Interferencia_1D.pdf", dpi=300, bbox_inches="tight")
plt.show()

files.download("Interferencia_1D.pdf")
```

```python
# =====================================================
# INTERFERENCIA 1D PARA DIFERENTES FASES
# =====================================================

x = np.linspace(0, 10, 500)

k = 2.0
omega = 2.0
t = 0.0

fases = [0, np.pi/2, np.pi]

plt.figure(figsize=(8,5))

for phi in fases:
    u = np.cos(k*x - omega*t) + np.cos(k*x - omega*t + phi)
    plt.plot(x, u, label=f"phi={round(phi,2)}")

plt.xlabel("x")
plt.ylabel("u(x,t)")
plt.title("Interferencia para diferentes fases")
plt.legend()

plt.tight_layout()
plt.savefig("Interferencia_1D_Fases.pdf", dpi=300, bbox_inches="tight")
plt.show()

files.download("Interferencia_1D_Fases.pdf")
```

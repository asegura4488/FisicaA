```python
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files

# ----------------------------------
# Definición de vectores
# ----------------------------------
A = np.array([3,2])
B = np.array([1,4])

# ----------------------------------
# Suma de vectores
# ----------------------------------
suma = A + B

print("Vector A =",A)
print("Vector B =",B)
print("A + B =",suma)

# ----------------------------------
# Crear figura
# ----------------------------------
plt.figure(figsize=(6,6))

# vectores desde el origen
plt.quiver(0,0,A[0],A[1], color='k',
           angles='xy',scale_units='xy',scale=1,label='A')

plt.quiver(0,0,B[0],B[1], color='b',
           angles='xy',scale_units='xy',scale=1,label='B')

# B desde la cabeza de A
plt.quiver(A[0],A[1],B[0],B[1],
           angles='xy',scale_units='xy',scale=1,
           color='green',label='B desde A')

# vector suma
plt.quiver(0,0,suma[0],suma[1],
           angles='xy',scale_units='xy',scale=1,
           color='red',label='A + B')

plt.xlim(-2,7)
plt.ylim(-2,7)

plt.axhline(0)
plt.axvline(0)

plt.xlabel("x")
plt.ylabel("y")
plt.title("Interpretación geométrica de la suma de vectores")

plt.legend()

# guardar figura
plt.savefig("Suma_vectores.pdf",dpi=300,bbox_inches='tight')

plt.show()

# descargar automáticamente
files.download("Suma_vectores.pdf")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files

# ----------------------------------
# Definición de vectores
# ----------------------------------
A = np.array([3, 2])
B = np.array([1, 4])

# ----------------------------------
# Resta de vectores
# ----------------------------------
resta = A - B

print("Vector A =", A)
print("Vector B =", B)
print("A - B =", resta)

# ----------------------------------
# Crear figura
# ----------------------------------
plt.figure(figsize=(6, 6))

# Vectores desde el origen
plt.quiver(0, 0, A[0], A[1], color='k',
           angles='xy', scale_units='xy', scale=1, label='A')
plt.quiver(0, 0, B[0], B[1], color='b',
           angles='xy', scale_units='xy', scale=1, label='B')

# Vector A-B dibujado desde la punta de B
plt.quiver(B[0], B[1], resta[0], resta[1],
           angles='xy', scale_units='xy', scale=1,
           color='red', label='A-B')

plt.xlim(-2, 6)
plt.ylim(-1, 6)

plt.axhline(0)
plt.axvline(0)

plt.xlabel("x")
plt.ylabel("y")
plt.title("Interpretación geométrica de la resta de vectores")
plt.legend()

# ----------------------------------
# Guardar figura
# ----------------------------------
plt.savefig("Resta_vectores.pdf", dpi=300, bbox_inches='tight')

plt.show()

# ----------------------------------
# Descargar automáticamente
# ----------------------------------
files.download("Resta_vectores.pdf")
```

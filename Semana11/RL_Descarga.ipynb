```python
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files

# =====================================================
# CIRCUITO RL: DESCARGA DE LA CORRIENTE
# =====================================================

# Parametros fisicos
R = 1.0
L = 1.0
I0 = 5.0   # corriente inicial

tau = L / R

# Tiempo
t_max = 5.0 * tau
t = np.linspace(0, t_max, 1000)

# Solucion analitica
I = I0 * np.exp(-t / tau)

# Voltajes
VR = R * I
VL = -VR

# Energia
U = 0.5 * L * I**2

# -----------------------------------------------------
# Corriente
# -----------------------------------------------------
plt.figure(figsize=(8, 5))
plt.plot(t, I, label='I(t)')
plt.axvline(tau, linestyle=':', label=r'$\tau=L/R$')
plt.xlabel('t')
plt.ylabel('Corriente')
plt.title('Descarga de corriente en un circuito RL')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.savefig("RL_Descarga_Corriente.pdf", dpi=300, bbox_inches="tight")
plt.show()

files.download("RL_Descarga_Corriente.pdf")

# -----------------------------------------------------
# Voltajes
# -----------------------------------------------------
plt.figure(figsize=(8, 5))
plt.plot(t, VR, label=r'$V_R(t)$')
plt.plot(t, VL, label=r'$V_L(t)$')
plt.axvline(tau, linestyle=':', label=r'$\tau=L/R$')
plt.xlabel('t')
plt.ylabel('Voltaje')
plt.title('Voltajes durante la descarga en un circuito RL')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.savefig("RL_Descarga_Voltajes.pdf", dpi=300, bbox_inches="tight")
plt.show()

files.download("RL_Descarga_Voltajes.pdf")

# -----------------------------------------------------
# Energia
# -----------------------------------------------------
plt.figure(figsize=(8, 5))
plt.plot(t, U, label='U(t)')
plt.axvline(tau, linestyle=':', label=r'$\tau=L/R$')
plt.xlabel('t')
plt.ylabel('Energia')
plt.title('Energia durante la descarga del inductor')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.savefig("RL_Descarga_Energia.pdf", dpi=300, bbox_inches="tight")
plt.show()

files.download("RL_Descarga_Energia.pdf")

print("===================================")
print("CIRCUITO RL: DESCARGA")
print("===================================")
print(f"R = {R}")
print(f"L = {L}")
print(f"I0 = {I0}")
print(f"tau = {tau}")
print(f"Energia inicial = {0.5 * L * I0**2:.4f} J")
```

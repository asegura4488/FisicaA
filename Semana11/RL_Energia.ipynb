```python
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files

# =====================================================
# CIRCUITO RL: ENERGIA ALMACENADA EN EL INDUCTOR
# =====================================================

# Parametros fisicos
R = 1.0
L = 1.0
V0 = 5.0

tau = L / R

# Tiempo
t_max = 5.0 * tau
t = np.linspace(0, t_max, 1000)

# Corriente
I = (V0 / R) * (1.0 - np.exp(-t / tau))

# Energia magnetica almacenada
U = 0.5 * L * I**2

# Potencia en la resistencia
PR = R * I**2

# -----------------------------------------------------
# Grafica de la energia
# -----------------------------------------------------
plt.figure(figsize=(8, 5))
plt.plot(t, U, label='U(t)')
plt.axvline(tau, linestyle=':', label=r'$\tau=L/R$')
plt.xlabel('t')
plt.ylabel('Energia')
plt.title('Energia almacenada en el inductor')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.savefig("RL_Energia.pdf", dpi=300, bbox_inches="tight")
plt.show()

files.download("RL_Energia.pdf")

# -----------------------------------------------------
# Grafica de la potencia disipada
# -----------------------------------------------------
plt.figure(figsize=(8, 5))
plt.plot(t, PR, label=r'$P_R(t)=RI^2$')
plt.axvline(tau, linestyle=':', label=r'$\tau=L/R$')
plt.xlabel('t')
plt.ylabel('Potencia')
plt.title('Potencia disipada en la resistencia')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.savefig("RL_Potencia_Resistencia.pdf", dpi=300, bbox_inches="tight")
plt.show()

files.download("RL_Potencia_Resistencia.pdf")

print("===================================")
print("CIRCUITO RL: ENERGIA")
print("===================================")
print(f"R = {R}")
print(f"L = {L}")
print(f"V0 = {V0}")
print(f"Energia final = {0.5 * L * (V0/R)**2:.4f} J")
```

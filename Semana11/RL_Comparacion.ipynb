```python
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files

# =====================================================
# COMPARACION DE CURVAS DE CORRIENTE PARA DISTINTOS RL
# =====================================================

V0 = 5.0
t = np.linspace(0, 10, 1000)

# Lista de parametros (R, L)
parametros = [
    (1.0, 1.0),
    (2.0, 1.0),
    (1.0, 2.0),
    (0.5, 1.0),
    (1.0, 0.5),
]

# -----------------------------------------------------
# Corriente para distintos valores de R y L
# -----------------------------------------------------
plt.figure(figsize=(8, 5))

for R, L in parametros:
    tau = L / R
    I = (V0 / R) * (1.0 - np.exp(-t / tau))
    etiqueta = f'R={R}, L={L}, tau={tau}'
    plt.plot(t, I, label=etiqueta)

plt.xlabel('t')
plt.ylabel('Corriente')
plt.title('Comparacion de corriente para distintos valores de R y L')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.savefig("RL_Comparacion_Corriente.pdf", dpi=300, bbox_inches="tight")
plt.show()

files.download("RL_Comparacion_Corriente.pdf")

# -----------------------------------------------------
# Voltaje en el inductor para distintos parametros
# -----------------------------------------------------
plt.figure(figsize=(8, 5))

for R, L in parametros:
    tau = L / R
    I = (V0 / R) * (1.0 - np.exp(-t / tau))
    VL = V0 - R * I
    etiqueta = f'R={R}, L={L}, tau={tau}'
    plt.plot(t, VL, label=etiqueta)

plt.xlabel('t')
plt.ylabel(r'$V_L(t)$')
plt.title('Voltaje en el inductor para distintos valores de R y L')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.savefig("RL_Comparacion_Inductor.pdf", dpi=300, bbox_inches="tight")
plt.show()

files.download("RL_Comparacion_Inductor.pdf")
```

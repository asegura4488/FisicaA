```python
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files

# =====================================================
# CIRCUITO RL: METODO DE EULER
# =====================================================

# Parametros fisicos
R = 1.0
L = 1.0
V0 = 5.0

tau = L / R

# Tiempo de simulacion
t_max = 5.0 * tau
dt = 0.01
t = np.arange(0, t_max + dt, dt)

# Arreglos
I_num = np.zeros_like(t)
VR_num = np.zeros_like(t)
VL_num = np.zeros_like(t)

# Condicion inicial
I_num[0] = 0.0

# Metodo de Euler
for n in range(len(t) - 1):
    dI_dt = (V0 - R * I_num[n]) / L
    I_num[n + 1] = I_num[n] + dt * dI_dt

# Voltajes
for n in range(len(t)):
    VR_num[n] = R * I_num[n]
    VL_num[n] = V0 - VR_num[n]

# Solucion analitica
I_ana = (V0 / R) * (1.0 - np.exp(-t / tau))

# -----------------------------------------------------
# Grafica de la corriente
# -----------------------------------------------------
plt.figure(figsize=(8, 5))
plt.plot(t, I_num, label='Numerico (Euler)')
plt.plot(t, I_ana, '--', label='Analitico')
plt.axvline(tau, linestyle=':', label=r'$\tau=L/R$')
plt.xlabel('t')
plt.ylabel('Corriente')
plt.title('Circuito RL: comparacion Euler vs solucion analitica')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.savefig("RL_Euler_Corriente.pdf", dpi=300, bbox_inches="tight")
plt.show()

files.download("RL_Euler_Corriente.pdf")

# -----------------------------------------------------
# Grafica de voltajes
# -----------------------------------------------------
plt.figure(figsize=(8, 5))
plt.plot(t, VR_num, label=r'$V_R(t)$')
plt.plot(t, VL_num, label=r'$V_L(t)$')
plt.axvline(tau, linestyle=':', label=r'$\tau=L/R$')
plt.xlabel('t')
plt.ylabel('Voltaje')
plt.title('Voltajes numericos en el circuito RL')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.savefig("RL_Euler_Voltajes.pdf", dpi=300, bbox_inches="tight")
plt.show()

files.download("RL_Euler_Voltajes.pdf")

print("===================================")
print("CIRCUITO RL: METODO DE EULER")
print("===================================")
print(f"R = {R}")
print(f"L = {L}")
print(f"V0 = {V0}")
print(f"tau = {tau}")
print(f"dt = {dt}")
```

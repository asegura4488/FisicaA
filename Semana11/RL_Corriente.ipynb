```python
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files

# =====================================================
# CIRCUITO RL: CRECIMIENTO DE LA CORRIENTE
# SOLUCION ANALITICA
# =====================================================

# Parametros fisicos
R = 1.0       # ohmios
L = 1.0       # henrios
V0 = 5.0      # voltios

# Constante de tiempo
tau = L / R

# Tiempo de simulacion
t_max = 5.0 * tau
t = np.linspace(0, t_max, 1000)

# Solucion analitica de la corriente
I = (V0 / R) * (1.0 - np.exp(-t / tau))

# Voltajes en la resistencia y el inductor
VR = R * I
VL = V0 - VR

# -----------------------------------------------------
# Grafica de la corriente
# -----------------------------------------------------
plt.figure(figsize=(8, 5))
plt.plot(t, I, label='I(t)')
plt.axvline(tau, linestyle=':', label=r'$\tau=L/R$')
plt.xlabel('t')
plt.ylabel('Corriente')
plt.title('Crecimiento de la corriente en un circuito RL')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.savefig("RL_Corriente.pdf", dpi=300, bbox_inches="tight")
plt.show()

files.download("RL_Corriente.pdf")

# -----------------------------------------------------
# Grafica de voltajes
# -----------------------------------------------------
plt.figure(figsize=(8, 5))
plt.plot(t, VR, label=r'$V_R(t)$')
plt.plot(t, VL, label=r'$V_L(t)$')
plt.axvline(tau, linestyle=':', label=r'$\tau=L/R$')
plt.xlabel('t')
plt.ylabel('Voltaje')
plt.title('Voltajes en la resistencia y el inductor')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.savefig("RL_Voltajes.pdf", dpi=300, bbox_inches="tight")
plt.show()

files.download("RL_Voltajes.pdf")

# Resumen numerico
print("===================================")
print("CIRCUITO RL: SOLUCION ANALITICA")
print("===================================")
print(f"R = {R}")
print(f"L = {L}")
print(f"V0 = {V0}")
print(f"tau = L/R = {tau}")
print(f"Corriente final = V0/R = {V0/R:.4f} A")
print(f"I(tau) = {(V0/R)*(1 - np.exp(-1)):.4f} A")
```

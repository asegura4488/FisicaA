```python
from google.colab import drive
drive.mount('/content/drive')
from google.colab import files

import numpy as np
import pandas as pd
import os

# =====================================================
# DATOS SINTETICOS: PARABOLICO CON FRICCION LINEAL
# =====================================================

# Carpeta de trabajo
carpeta = "/content/drive/MyDrive/Datos_Proyecto1"
os.makedirs(carpeta, exist_ok=True)

ruta = os.path.join(carpeta, "trayectoria_parabolica_friccion.csv")

# Parametros fisicos verdaderos
g = 9.81
v0x_real = 12.0
v0y_real = 10.0
tau_real = 1.8

# Tiempo
t = np.linspace(0, 2.0, 35)

# Modelo exacto
A = tau_real * (1 - np.exp(-t / tau_real))
x = v0x_real * A
y = tau_real * (v0y_real + g * tau_real) * (1 - np.exp(-t / tau_real)) - g * tau_real * t

# Ruido pequeno
rng = np.random.default_rng(42)
x_ruido = x + rng.normal(0, 0.05, size=len(x))
y_ruido = y + rng.normal(0, 0.05, size=len(y))

# Guardar CSV
df = pd.DataFrame({
    "time": t,
    "x": x_ruido,
    "y": y_ruido
})

df.to_csv(ruta, index=False)

print("Archivo guardado en:", ruta)
files.download(ruta)
```

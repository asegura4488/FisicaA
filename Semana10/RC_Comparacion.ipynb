```python
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files

# =====================================================
# COMPARACIÓN DE CURVAS DE CARGA PARA DISTINTOS RC
# =====================================================

V0 = 5.0
t = np.linspace(0, 10, 1000)

# Lista de parámetros (R, C)
parametros = [
    (1.0, 1.0),
    (2.0, 1.0),
    (1.0, 2.0),
    (0.5, 1.0),
    (1.0, 0.5),
]

# -----------------------------------------------------
# Comparación de voltajes
# -----------------------------------------------------
plt.figure(figsize=(8, 5))

for R, C in parametros:
    tau = R * C
    Vc = V0 * (1.0 - np.exp(-t / tau))
    etiqueta = f'R={R}, C={C}, tau={tau}'
    plt.plot(t, Vc, label=etiqueta)

plt.xlabel('t')
plt.ylabel(r'$V_C(t)$')
plt.title('Comparación de carga del condensador para distintos valores de RC')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.savefig("RC_Comparacion_Carga.pdf", dpi=300, bbox_inches="tight")
plt.show()
files.download("RC_Comparacion_Carga.pdf")

# -----------------------------------------------------
# Comparación de corrientes
# -----------------------------------------------------
plt.figure(figsize=(8, 5))

for R, C in parametros:
    tau = R * C
    I = (V0 / R) * np.exp(-t / tau)
    etiqueta = f'R={R}, C={C}, tau={tau}'
    plt.plot(t, I, label=etiqueta)

plt.xlabel('t')
plt.ylabel(r'$I(t)$')
plt.title('Comparación de corriente para distintos valores de RC')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.savefig("RC_Comparacion_Corriente.pdf", dpi=300, bbox_inches="tight")
plt.show()
files.download("RC_Comparacion_Corriente.pdf")
```

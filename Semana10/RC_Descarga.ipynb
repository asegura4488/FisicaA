```python
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files

# =====================================================
# CIRCUITO RC: DESCARGA DEL CONDENSADOR
# =====================================================

# Parámetros físicos
R = 1.0
C = 1.0
Vi = 5.0

tau = R * C

# Tiempo de simulación
t_max = 5.0 * tau
dt = 0.01
t = np.arange(0, t_max + dt, dt)

# Arreglos numéricos
Vc_num = np.zeros_like(t)
I_num = np.zeros_like(t)
U_num = np.zeros_like(t)

# Condición inicial
Vc_num[0] = Vi

# Método de Euler
for n in range(len(t) - 1):
    dVc_dt = -Vc_num[n] / (R * C)
    Vc_num[n + 1] = Vc_num[n] + dt * dVc_dt

# Corriente y energía
for n in range(len(t)):
    I_num[n] = -Vc_num[n] / R
    U_num[n] = 0.5 * C * Vc_num[n]**2

# Solución analítica
Vc_ana = Vi * np.exp(-t / tau)
I_ana = -(Vi / R) * np.exp(-t / tau)
U_ana = 0.5 * C * Vc_ana**2

# -----------------------------------------------------
# Gráfica 1: voltaje
# -----------------------------------------------------
plt.figure(figsize=(8, 5))
plt.plot(t, Vc_num, label='Numérico (Euler)')
plt.plot(t, Vc_ana, '--', label='Analítico')
plt.axvline(tau, linestyle=':', label=r'$\tau=RC$')
plt.xlabel('t')
plt.ylabel(r'$V_C(t)$')
plt.title('Descarga del condensador en un circuito RC')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.savefig("RC_Descarga_Voltaje.pdf", dpi=300, bbox_inches="tight")
plt.show()
files.download("RC_Descarga_Voltaje.pdf")

# -----------------------------------------------------
# Gráfica 2: corriente
# -----------------------------------------------------
plt.figure(figsize=(8, 5))
plt.plot(t, I_num, label='Numérico (Euler)')
plt.plot(t, I_ana, '--', label='Analítico')
plt.axvline(tau, linestyle=':', label=r'$\tau=RC$')
plt.xlabel('t')
plt.ylabel(r'$I(t)$')
plt.title('Corriente durante la descarga en un circuito RC')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.savefig("RC_Descarga_Corriente.pdf", dpi=300, bbox_inches="tight")
plt.show()
files.download("RC_Descarga_Corriente.pdf")

# -----------------------------------------------------
# Gráfica 3: energía
# -----------------------------------------------------
plt.figure(figsize=(8, 5))
plt.plot(t, U_num, label='Numérico (Euler)')
plt.plot(t, U_ana, '--', label='Analítico')
plt.axvline(tau, linestyle=':', label=r'$\tau=RC$')
plt.xlabel('t')
plt.ylabel(r'$U(t)$')
plt.title('Energía durante la descarga del condensador')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.savefig("RC_Descarga_Energia.pdf", dpi=300, bbox_inches="tight")
plt.show()
files.download("RC_Descarga_Energia.pdf")

print("===================================")
print("CIRCUITO RC: DESCARGA")
print("===================================")
print(f"R = {R}")
print(f"C = {C}")
print(f"Vi = {Vi}")
print(f"tau = RC = {tau}")
print(f"Vc(tau) analítico = {Vi * np.exp(-1):.4f} V")
print(f"I(0) = {-Vi/R:.4f} A")
print(f"Energía inicial = {0.5 * C * Vi**2:.4f} J")
```

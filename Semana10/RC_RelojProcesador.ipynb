```python
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files

# =====================================================
# LÍMITE DEL RELOJ DEL PROCESADOR BASADO EN RC
# =====================================================

# Parámetros base
R = 100.0
C = 1e-12
k = 5.0

tau = R * C
T_min = k * tau
f_max = 1.0 / T_min

print("===================================")
print("LÍMITE DEL RELOJ DEL PROCESADOR")
print("===================================")
print(f"R = {R:.3e} ohm")
print(f"C = {C:.3e} F")
print(f"tau = RC = {tau:.3e} s")
print(f"T_min = k*tau = {T_min:.3e} s")
print(f"f_max = 1/T_min = {f_max:.3e} Hz")
print(f"f_max = {f_max/1e9:.3f} GHz")

# -----------------------------------------------------
# Frecuencia máxima vs capacitancia
# -----------------------------------------------------
C_values = np.logspace(-15, -10, 200)
f_values = 1.0 / (k * R * C_values)

plt.figure(figsize=(8, 5))
plt.loglog(C_values, f_values)
plt.xlabel('Capacitancia C (F)')
plt.ylabel('Frecuencia máxima (Hz)')
plt.title('Frecuencia máxima vs capacitancia')
plt.grid(True, which='both')
plt.tight_layout()
plt.savefig("RC_Reloj_vs_C.pdf", dpi=300, bbox_inches="tight")
plt.show()
files.download("RC_Reloj_vs_C.pdf")

# -----------------------------------------------------
# Frecuencia máxima vs resistencia
# -----------------------------------------------------
R_values = np.logspace(0, 4, 200)
f_values_R = 1.0 / (k * R_values * C)

plt.figure(figsize=(8, 5))
plt.loglog(R_values, f_values_R)
plt.xlabel('Resistencia R (ohm)')
plt.ylabel('Frecuencia máxima (Hz)')
plt.title('Frecuencia máxima vs resistencia')
plt.grid(True, which='both')
plt.tight_layout()
plt.savefig("RC_Reloj_vs_R.pdf", dpi=300, bbox_inches="tight")
plt.show()
files.download("RC_Reloj_vs_R.pdf")
```
